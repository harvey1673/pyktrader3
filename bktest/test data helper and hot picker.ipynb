```python

def get_run_date():
    tday = datetime.date.today()
    now = datetime.datetime.now()
    if misc.is_workday(tday, 'CHN') and now.hour < 18:
        run_date = tday
    else:
        run_date = misc.day_shift(tday, '1b', misc.CHN_Holidays)
    return run_date
rdate = get_run_date()
rdate.strftime('%Y%m%d')
```
