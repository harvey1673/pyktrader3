```python
def generate_calendar_json(start_date, end_date, filename, hols = misc.CHN_Holidays, date_for = "%Y%m%d"):
    d = start_date
    dlist = []
    while d <= end_date:
        if misc.is_workday(d, calendar = 'CHN'):
            dlist.append(datetime.datetime.strftime(d, date_for))
        d = d + datetime.timedelta(days=1)
    
    outfile = open(filename, "w")
    outfile.write(json.dumps(dlist, indent=4))
    outfile.close()
    

generate_calendar_json(datetime.date(2021,1,1), datetime.date(2021,12,31), "C:\\dev\\data\\calendar.json")
```
